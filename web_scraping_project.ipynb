```python
from bs4 import BeautifulSoup
import requests
```

```python
url= "https://edgar.jrc.ec.europa.eu/report_2024"
page= requests.get(url)
soup= BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
table= soup.find('table')
```

```python
print(table)
```

```python
soup.find('th')
```

```python
table_titles=soup.find_all('th')
```

```python
emission_titles= [title.text for title in table_titles][0:9]
print(emission_titles)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns=emission_titles)
df
```

```python
column_data=soup.find_all('tr')
```

```python
for row in column_data[2:]:
    
    row_data=row.find_all('td')
    emission_data= [data.text for data in row_data]
    print(emission_data)
```

```python
for row in column_data[2:]:
    
    row_data=row.find_all('td')
    emission_data= [data.text for data in row_data]
    length=len(df)
    df.loc[length]=emission_data
```
